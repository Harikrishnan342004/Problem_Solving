```python
def matchingStrings(stringList, queries):
    # Write your code here
    target = []
    for p_element in queries:
        count = 0
        for stringList_ele in stringList:
            if(p_element == stringList_ele):
                count += 1
        target.append(count)
    return target
stringList = ['ab','ab','abc']
queries = ['ab','abc','bc']
result = matchingStrings(stringList, queries)
print(result)
```
