```python
#reverse the string...
def reverse(string):
    empty = ""
    for i in string:
        #print(i)
        empty = i + empty
    return empty
string = "Hari"
Output = reverse(string)
print(Output)
```

```python
#reverse the string...
def reverse(string):
    empty = ""
    for i in range(len(string)-1,-1,  -1):
        #print(i)
        empty += string[i] 
    return empty
string = "Hari"
Output = reverse(string)
print(Output)
```

```python
#reverse the string...
def reverse(string):
    empty = " "
    empty += string[::-1]
    return empty
    
string = "Hari"
Output = reverse(string)
print(Output)
```

```python
# remove Duplicates From the String By Using not in
def Remove_duplicates(Input):
    result = ""
    for char in Input:
        if char not in result:
            result += char

    return result
Input = "Harikrishnan"
output = Remove_duplicates(Input)
print(output)
```

```python
# Remove Duplicates By using Set method

def Remove_Dupliates(Input):
    result = ""
    removing_dup = set(Input)
    for char in removing_dup:
        result += char
    return result
Input = "Harikrishnan"
output = Remove_duplicates(Input)
print(output)
```

```python
#Remove Consiquence Duplicate value from the strings   #Array Method
def Remove_Duplicates(Input):
    result = " "
    length = len(Input)
    for i in range(length):
        if(  i+1 < length and Input[i] == Input[i + 1] ):
            continue
        result += Input[i]
    return result

Input = "aabbccbaab"
Output = Remove_Duplicates(Input)
print(Output)
```

```python
#Remove vowels from string...
def remove_vowels(Input):
    result = ""
    vowels = "aeiou"

    for char in Input: 
        if(char not in vowels):
            result += char
    return result
Input = "aabbccbaab"
Output = remove_vowels(Input)
print(Output)
```

```python
#find the first unique character  in the string leetcode
def unique(s):
    empty = {}
    for char in s:
        if char in empty:
            empty[char] += 1
        else:
            empty[char] = 0
    
    for i in range(len(s)):
        if( empty[s[i]] == 0):
            return s[i]

    return empty
s = "leetcode"
Output = unique(s)
print(Output)
```
