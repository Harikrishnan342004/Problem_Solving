```python
# Next Smallest Palindrome.
# Given a number as an input, find the following smallest palindrome number. 
# For simplicity, assume the input value will not exceed 1 million. The next palindrome may be greater than 1 million.
# Constraints
# The input variable n is an integer.
# The input value n will not exceed 1,000,000. The resulting palindrome may be greater than this limit.

def is_palindrome(num):
    return str(num) == str(num)[::-1]

def next_palindrome(n):
    n += 1
    print(f"Checking outer {n} for palindrome...")

    while not is_palindrome(n):
        n += 1
        print(f"Checking inner {n} for palindrome...")

    return n

n = 12
result = next_palindrome(n) 
print(result)
```
