```python
def simpleArraySum(arr):
    target = 0
    for i in arr:
        target += i
    return target
arr = [1,2,3]
print(simpleArraySum(arr))
```

```python
# Arrays - DS
# reverse 

def reverseArray(a):
    result = [ ]
    for i in a:
        result = [i] + result
    return result
a = [1,2,3]
print(reverseArray(a))
```

```python
def reverseArray(a):
    a.reverse()
    return a
a = [1,2,3]
print(reverseArray(a))
```

```python
def reverseArray(a):
    b = a[::-1]
    return b
a = [1,2,3]
print(reverseArray(a))
```
