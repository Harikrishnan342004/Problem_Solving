```python
# 121. Best Time to Buy and Sell Stock

class Solution:
    def maxProfit(self, prices: List[int]) -> int:
        highest_provit_value = 0

        for i in range(0 , len(prices)):
            for j in range(i+1 , len(prices)):
                if(prices[j]-prices[i] > highest_provit_value):
                    highest_provit_value = prices[j]-prices[i]
        return highest_provit_value
prices = [7,1,5,3,6,4]
print(Solution().maxProfit(prices))
        
```

```python
# 217. Contains Duplicate
# Input: nums = [1,2,3,1]
# Output: true
# Explanation:
# The element 1 occurs at the indices 0 and 3.

class Solution:
    def containsDuplicate(self, nums: List[int]) -> bool:
        return len(nums) != len(set(nums))

nums = [1,2,3,1]
print(Solution().containsDuplicate(nums))
        
```

```python
# 238. Product of Array Except Self.

class Solution:
    def productExceptSelf(self, nums: List[int]) -> List[int]:
        result = []
        for i in range(len(nums)):
            product = 1
            for j in range(len(nums)):
                if(i != j):
                    product *= nums[j]
            result.append(product)
        return result
nums = [1,2,3,4]
print(Solution().productExceptSelf(nums))      

```

```python
def simpleArraySum(arr):
    target = 0
    for i in arr:
        target += i
    return target
arr = [1,2,3]
print(simpleArraySum(arr))
```

```python
# Maximum Sum

# Write a function that takes an array of integers and returns the maximum sum of any contiguous subarray.
# The subarray must contain at least one element.

# "Contiguous" means the subarray consists of elements that are next to each other in the original array 
# (i.e., they appear sequentially without gaps).

# Constraints
# The input variable must be a list of integers.
# The list must contain at least one element: 1 ≤ n ≤ 10^5.
# The elements in the list can be positive, negative, or zero: -10^4 ≤ arr[i] ≤ 10^4.
# The function must return a single integer, representing the maximum sum.
# The subarray elements must be contiguous (no reordering allowed).
# If all numbers in the list are negative, the function must return the least negative number (i.e., the maximum element in the list).
# The algorithm must be efficient (O(n)) since n can be large.
# More about this question
# Hints
# Test Case #1
# Input: {"nums": [1, -2, 3, 4, -1, 2, 1, -5, 4]}
# Output: 9
# Description: This test case examines the function’s ability to find the maximum sum in a list with both positive and negative numbers.
# Test Case #2
# Input: {"nums": [-3, -5, -2, -8, -1]}
# Output: -1
# Description: Tests the function’s handling of an all-negative list, where the maximum subarray sum is the least negative number.
# Test Case #3
# Input: {"nums": [5, -1, -2, 7, -3, 4]}
# Output: 10
# Description: Verifies that the function correctly identifies the maximum subarray in a case where a mix of losses and gains occurs.

def max_sub_sum(nums):
    result = 0
    for i in nums:
        result += i
        if result < 0:
            result = 0
    return result
print(max_sub_sum([1, -2, 3, 4, -1, 2, 1, -5, 4]))   #Output is 9
print()


```
