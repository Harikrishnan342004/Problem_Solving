```python
"""  Using Recursion reverse the string such as

Eg 1: Input: one two three
      Output: three two one
Eg 2: Input: I love india
      Output: india love I """

def reverse_string(Input):
    words = []
    word = " "
    
    for char in Input:
        
        if (char != " "):
            
            word += char
        else:
            if word:
                words.insert(0 , word)
                word =" "
                
    if word:
        words.insert(0 , word)
        
    return " ". join(words)

Input = "santhiya Hari krishnan"
output = reverse_string(Input)
print(output)
```

```python

def remove_duplicates(input_string):
    # Create an empty set to store encountered characters
    seen_chars = set()

    # Initialize an empty string to store the result
    result = ""

    for char in input_string:
        # If the character has not been encountered, add it to the result and the set
        if char not in seen_chars:
            result += char
            seen_chars.add(char)

    return result

# Example usage:
testcases = ["Java1234", "Python1223", "aBuzZ9900"]

for test in testcases:
    result = remove_duplicates(test)
    print(f"Input: {test}")
    print(f"Output: {result}\n")
```

```python
class Train:           #Two classs are definded in this program one is class Train ......... 
    #Tha contain capacity , passengers, waiting_list
    def __init__(self, capacity = 50):
        self.capacity = capacity
        self.passengers = []
        self.waiting_list = []
    
    def book_ticket(self, passenger):
        if ( len(self.passengers) < self.capacity):
            self.passengers.append(passenger)
            print("Ticket Booked")
        else:
            self.waiting_list.append(passenger)
            print("Train is full. Added to waiting list")
    
    def cancel_ticket(self, passenger):
        if passenger in self.passengers:
            self.passengers.remove(passenger)
            print("Cannceled")
        elif passenger in self.waiting_list:
            self.passengers.remove(passenger)
            print("Cannceled from waiting list")
        else:
            print("Passenger not found")


    def print_pnr_details(self):
        print("Passengers with confirmed tickests:")
        for passenger in self.passengers:
            print(passenger.name)
        print("Passenger on waiting list:")
        for Passenger in self.waiting_list:
            print(passenger.name)


    
class Passenger:    # Another class is Passengers that contain passenger name Attributes
    def __init__(self, name):
        self.name = name


def main():
    train = Train()
    print("Welcome to Ticket Reservation System\n")
    
    while True:
        print("\n1. Book Ticket \n 2.Cancel Ticket \n 3.Print PNR Details \n 4.Exit")
        choice = input("Enter Passenger choice: ")

        if choice == "1":
            name = input("Enter passenger name: ")
            passenger = Passenger(name)
            train.book_ticket(passenger)
        
        elif choice == "2":
            name = input("enter passenger name: ")
            passenger = Passenger(name)
            train.cancel_ticket(passenger)
        elif choice == "3":
            train.print_pnr_details()
        elif choice == "4":
            print("Exiting...")
        else:
            print("Invalid choice. Please enter a valid option")



if __name__ == "__main__":
    main()
```

```python
#combination sum...
def combination_sum(arr, target):
    def backtracking( start , path , target ):
        if(target == 0 ):
            result.append(path)
        if( target < 0 ):
            return 
        for i in range(start , len(arr)):
            backtracking( i , path+[arr[i]] , target-arr[i] )
    result = []
    arr.sort()
    backtracking(0, [] , target)
    return result
arr = [2,3,5]
target = 8
output = combination_sum(arr, target)
print(output)
```

```python
#Spiral Matrix
def spiral_Matrix(matrix):
    if not matrix:
        return []
    m = len(matrix)
    n = len(matrix[0])
    ans = []
    top = 0
    left = 0
    bottom = m-1
    right = n-1
    while len(ans) < m * n:
        j = left
        while( j <= right and len(matrix) < m * n ):
            ans.append(matrix[top][j])
            j += 1
        i = top + 1
        while i<= bottom-1 and len(matrix) < m * n :
            ans.append(matrix[i][right])
            i += 1
        j = right
        while  j >= left and len(matrix)< m * n:
            ans.append(matrix[bottom][j] )
            j -= 1
            
            
        i = bottom -1
        while ( i >= top+1 and len(matrix) < m * n ):
            ans.append(matrix[i][left])
            i -= 1
        
        top += 1
        left += 1
        bottom -= 1
        right -= 1
    return ans
            
        
matrix = [[1,2,3],[4,5,6],[7,8,9]]
output = spiral_Matrix(matrix)
print(output)
```

```python
#Calculate forex profit
def calculate_forex_profit(price):
    length = len(price)
    Total_profit = 0
    buying_hour = -1
    
    for i in range(length-1):
        if ( buying_hour == -1 or price[i] < price[buying_hour] ):
            buying_hour = i
        
        if(price[i+1] > price[i]):
            Total_profit += price[i+1] - price[buying_hour]
            print(f"buying hour: {buying_hour +1}, selling hour: {buying_hour +2} ")
            buying_hour = -1
        
    return Total_profit
price  = [5,5,10,8,7,6,4,5,2,6,1,10]
output =  calculate_forex_profit(price)
print(output)
```

```python
#............4.........
def findCombination(digit, i=0, s=''):
    if not digit:
        return []
    if i == len(digit):
        return [s]
    
    alpha = 'ABCDEFGHIJKLMNOPQRSTUVWXYZ'
    total = 0
    combinations = []
    
    # process the next two digits (i'th and (i+1)'th) 
    for j in range(i, min(i + 1, len(digit) - 1) + 1):
        total = (total * 10) + digit[j] 
 
        # if a valid character can be formed by taking one or both digits,
        # append it to the output and recur for the remaining digits
        if 0 < total <= 26:
            combinations.extend(findCombination(digit, j + 1, s + alpha[total - 1]))
            
    return combinations

digit = [2,6,1,8]
output = findCombination(digit)
print(output)
```

```python
def fact(n):
    if n <= 1:
        return 1
    
    return n * fact(n-1)

def findRank(s):
    
    length = len(s)
    rank = 1
    
    for i in range(length):
       
        count = 0
        
        for j in range(i+1, length):
            
            if(s[i] > s[j]):
                count += 1
        rank += count  * fact(length-i-1)
    
    return rank


def main():
    s = "string"
    print(findRank(s))
    return 0
main()
```

```python
def reverse_string_with_punctuation(Input):  
    #coverting the string into list
    string_list = list(Input)
    
    left, right = 0 , len(Input)-1
    
    while left< right:
        
        if not string_list[left].isalnum():
            left += 1
        elif not string_list[right].isalnum():
            right -= 1
        else:
            string_list[left],string_list[right] = string_list[right], string_list[left]
            left += 1
            right -= 1
        
    return ''.join(string_list)
    
    
    
Input = "house! no: 12"
output = reverse_string_with_punctuation(Input)
print(output)
```

```python
#Reversu the vowels of the string..........

def reverse(Input):
    string_to_list = list(Input)
    i = 0
    j = len(Input) - 1
    vowels = ["a","e" , "i" , "o" , "u", "A" , "E", "I" , "O" , "U"]
    
    while i<j:
        if(string_to_list[i] in vowels and string_to_list[j] in vowels ):
            string_to_list[i],string_to_list[j] = string_to_list[j],string_to_list[i]
            i += 1
            j -= 1
        elif(string_to_list[i] not in vowels):
            i += 1
        elif(string_to_list[j] not in vowels):
            j -= 1
        else:
            i += 1
            j -= 1
    return "".join(string_to_list)

Input = "Hari"
output = reverse(Input)
print(output)
```

```python
#Median of two shorted array...........

def Median(Input1, Input2):
    merged = sorted(Input1 + Input2)
    length = len(merged)
    
    if(length % 2 == 0):
        return (merged[length // 2] -1  +  merged[length // 2]) / 2
    else:
        return merged[length // 2]


Input1 = [1,2,3]
Input2 = [4,5]
output = Median(Input1, Input2)
print(output)
```

```python
#Rotate Matrix.......

def rotate(matrix):
    length = len(matrix)
    
    for i in range(length):
        for j in range(1+i, length):
            matrix[i][j],matrix[j][i] = matrix[j][i],matrix[i][j]
            
    for i in range(length):
        matrix[i].reverse()
    return matrix


matrix = [
 [1 , 2 , 3],
 [4 , 5 , 6],
 [7 , 8 , 9]
]
output = rotate(matrix)
print(output)
```

```python
#ways to target...

def countWays(A, B , C , Target):
    if Target<0:
        return 0
    if Target == 0:
        return 1
    
    Way_with_A = countWays(A, B, C , Target-A)
    Way_with_B = countWays(A, B , C, Target-B)
    Way_with_C = countWays(A, B , C , Target-C)
    
    Total_ways = Way_with_A +  Way_with_B +  Way_with_C
    
    return Total_ways
    
    
A = 3
B = 2
C = 1
Target = 2
result = countWays(A, B , C ,Target)
print(result)
```

```python
#beautiful pole .................

# s , we need to find no. of beautiful poles. (Pole is beautiful if it's lesser than or equal to right pole)

# Ex. Inp - [1,1,3,1]
#       Out - 5
# Reason :
#  pole[0] is beautiful when it's compared with [1] , [2] and [3]

# pole[1] is beautiful when it's compared with [2] and [3]

# pole[2] is not beautiful because all right pole are shorter

# pole[3] is not beautiful because it doesn't contain any right pole
def max_equal_height(n1, n2, n3, poles):
    heights = [sum(pole) for pole in poles]
    
    
    while len(set(heights)) > 1:
        max_height = max(heights)
        for i in range(3):
            if heights[i] == max_height:
                heights[i] -= poles[i].pop(0)
    
    return heights[0]

# Given input
n1, n2, n3 = 5, 3, 4
pole1 = [3, 2, 1, 1, 1]
pole2 = [4, 3, 2]
pole3 = [1, 1, 4, 1]

poles = [pole1, pole2, pole3]

# Calculate and print the maximum equal height
print(max_equal_height(n1, n2, n3, poles))


```

```python
#Encrypt_message.......
def encrypt_message(Message , shift):
    encrypted_message = " "
    
    for char in Message:
       
        if char.isalpha():
            shift_amount = shift % 26
            if char.islower():
                encrypted_char = chr( (ord(char) - ord("a") + shift_amount ) % 26 + ord("a") )
            else:
                encrypted_char = chr(  (ord(char) - ord("A") + shift_amount ) % 26 + ord("A"))
        elif (char.isdigit()):
            shift_amount  = shift % 10
            encrypted_char = chr((ord(char) - ord("0") + shift_amount) % 10 + ord("0"))
         
        else:
            encrypted_char = char
        
        encrypted_message += encrypted_char
        
    return encrypted_message
            
            
Message = "Wake me early at 5 a.m"
shift = 3
output = encrypt_message(Message , shift)
print("Enctypt_Output:", output)
```

```python
def longest_consecutive_number(nums):
    if not nums:
        return 0
    
    max_length = 0
    num_set = set(nums)
    
    for i in num_set:
        current_num = i
        current_length = 1
        if i - 1 not in num_set:
            while current_num + 1 in num_set:
                current_num += 1
                current_length += 1
        
            max_length = max(max_length, current_length)
    return max_length

nums = [1, 2, 3, 4, 5, 6, 7,]
output = longest_consecutive_number(nums)
print(output)
```
