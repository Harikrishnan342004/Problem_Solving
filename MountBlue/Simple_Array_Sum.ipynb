```python
def simpleArraySum(arr):
    target = 0
    for i in arr:
        target += i
    return target
arr = [1,2,3]
print(simpleArraySum(arr))
```

```python
# Sales by Match
def simpleArray(ar):
    result = {}
    pairs = 0
    for i in ar:
        if(i in result):
            result[i] += 1
        else:
            result[i] = 1
    print(result)
    for j in result:
        pairs += result[j]//2
    return pairs

ar = [10, 20, 20, 10, 10, 30, 50, 10, 20]
print(simpleArray(ar))
```
