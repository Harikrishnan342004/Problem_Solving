```python
def simpleArraySum(arr):
    target = 0
    for i in arr:
        target += i
    return target
arr = [1,2,3]
print(simpleArraySum(arr))
```

```python
# Sales by Match
def simpleArray(ar):
    result = {}
    pairs = 0
    for i in ar:
        if(i in result):
            result[i] += 1
        else:
            result[i] = 1
    print(result)
    for j in result:
        pairs += result[j]//2
    return pairs

ar = [10, 20, 20, 10, 10, 30, 50, 10, 20]
print(simpleArray(ar))
```

```python
# Breaking the Records



def breakingRecords(arr):
    min_score = arr[0]
    max_score = arr[0]
    min_count = 0
    max_count = 0

    for ele in arr[1:]:
        if ele < min_score:
            min_score = ele
            min_count += 1
        elif ele > max_score:
            max_score = ele
            max_count += 1

    return [max_count, min_count]

arr = [3, 4, 21, 36, 10, 28, 35, 5, 24, 42]
print(breakingRecords(arr))
```
