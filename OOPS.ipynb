```python
#Polimorpshim
class Car:
    def fuel_type(self):
        pass
class ECar(Car):
    def fuel(self):
        return "Electricity"
class GCar(Car):
    def fuel(self):
        return "gas"
def car_fuel(car):
    print(f"The car runs on:{car.fuel()}")

electric_car = ECar()
gasoline_car = GCar()

car_fuel(electric_car)
car_fuel(gasoline_car)
```

```python
#Inheritance
#parent class
class Car:
    def __init__(self, make , model):
        self.make = make
        self.model = model
    def display_info(self):
        print(f"Car: {self.make} {self.model}")
#child class
class ElectricCar(Car):
    def __init__(self, make , model , battery):
        # Initialize attributes of the parent class
        super().__init__(make, model)

        self.battery = battery
    def display_batter_info(self):
        print(f"Bater size: {self.battery}")
my_electric_car = ElectricCar("Tesla", "Model 3", 75)

my_electric_car.display_batter_info()
my_electric_car.display_info()
```

```python
#Encapsulation
class Smartphone:
    def __init__(self, brand , os):
        self.brand = brand
        self.os = os
    def display(self):
        print(f"display OS: {self.brand}, {self.os}")
iphone = Smartphone("Apple", "ios 17")
iphone.display()
```
