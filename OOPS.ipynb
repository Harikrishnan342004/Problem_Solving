```python
#Class and Objects

#Defining the Car Class

class Car:
    def __init__(self, model , year):
        self.model = model
        self.year = year

    def display_info(self):
        print(f"Car Details: {self.year}  {self.model}")

# Creating an object of the Car class
my_car = Car("Toyata", 2020)

my_car.display_info()
```

```python
#Polimorpshim
class Car:
    def fuel_type(self):
        pass
class ECar(Car):
    def fuel(self):
        return "Electricity"
class GCar(Car):
    def fuel(self):
        return "gas"
def car_fuel(car):
    print(f"The car runs on:{car.fuel()}")

electric_car = ECar()
gasoline_car = GCar()

car_fuel(electric_car)
car_fuel(gasoline_car)
```
