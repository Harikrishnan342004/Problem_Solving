```python
#sum Find the sum of all numbers from 1 to N that are divisible by both 2 and 5.
def sum(Input):
    result = 0
    for i in range(1, Input+1):
        if( i % 2 == 0 and i % 5 == 0):
            #print(i)
            result = result +  i
    return result
Input = 20
Output = sum(Input)
print(Output)
```

```python
def solve(input1):
    s = list(str(input1))  # convert to list for easy manipulation
    
    # Step 1: find smallest digit and its first occurrence index
    min_digit = min(s)
    min_index = s.index(min_digit)
    
    # Step 2: move that digit to the front
    s.pop(min_index)
    s.insert(0, min_digit)
    
    # Step 3: form the number
    result = int("".join(s))
    
    return [result, min_index, 0]


# Example
print(solve(828))
```
