```python
#find the first unique character  in the string leetcode
def unique(s):
    empty = {}
    for char in s:
        if char in empty:
            empty[char] += 1
        else:
            empty[char] = 0
    
    for i in range(len(s)):
        if( empty[s[i]] == 0):
            return s[i]

    return empty
s = "eetcode"
Output = unique(s)
print(Output)
```
