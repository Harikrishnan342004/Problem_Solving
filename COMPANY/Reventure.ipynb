```python
def  Average(n):
    add = 0
    for i in range(0, n+1):
        add += i
    result = add / (n+1)
    return result
n = 4
Input = Average(n)
print(Input)
```

```python
def count_average(input1):
    target = {}
    for char in  input1:
        if( char in target):
            target[char] += 1
        else:
            target[char] = 1
    return target

input1 = "Hello"
Output = count_average(input1)
print(Output)
```
