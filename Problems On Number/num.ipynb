```python
#sum Find the sum of all numbers from 1 to N that are divisible by both 2 and 5.
def sum(Input):
    result = 0
    for i in range(1, Input+1):
        if( i % 2 == 0 and i % 5 == 0):
            #print(i)
            result = result +  i
    return result
Input = 20
Output = sum(Input)
print(Output)
```
